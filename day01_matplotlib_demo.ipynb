### 完善原始折线图 — 给图形添加辅助功能

**画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度**

```python
import random

# 1.准备数据 x y
x = range(60)
y_shanghai = [random.uniform(17, 18) for i in x]

# 2..创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3..绘制图像
# 设置中文字体
plt.rc('font', family='simsun', size=13)
plt.plot(x, y_shanghai)

# 修改x，y刻度
# 准备x的刻度说明
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(0, 40, 2))

# 添加网格显示
plt.grid(linestyle="--", alpha=0.3)
# 添加描述信息
plt.xlabel("时间变化")
plt.ylabel("温度变化")
plt.title("某城市11-12点的温度变化")

# 4.显示图像
plt.show()
```

### 完善原始折线图2（图像层）

**需求：再添加一个城市的温度变化**

```python
# 收集到北京当天温度变化情况，温度在1度到3度。
# 怎么去添加另一个在同一坐标系当中的不同图形，其实很简单只需要再次plot即可，但是需要区分线条，如下显示
```

```python
# 1.准备数据 x y
y_beijin = [random.uniform(1, 3) for i in x]

# 2.创建画布
plt.figure(figsize=(20, 8), dpi=80)
# 3.绘制图像
plt.plot(x, y_shanghai, color="b", linestyle="--", label="上海")
plt.plot(x, y_beijin, color="r", linestyle=":", label="北京")


# 显示图例
plt.legend(loc="best")

# 修改x，y刻度
# 准备x的刻度说明
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(0, 30, 2))

# 添加网格显示
plt.grid(linestyle="--", alpha=0.3)
# 添加描述信息
plt.xlabel("时间变化")
plt.ylabel("温度变化")
plt.title("上海，北京11-12点的温度变化")
# 4.显示图像
plt.show()
```

**多个坐标系显示— plt.subplots(面向对象的画图方法)**

```python
# 1.准备数据 x y
y_beijin = [random.uniform(1, 3) for i in x]

# 2.创建画布
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=80)
# 3.绘制图像
plt.rc('font', family='simsun', size=15)
axes[0].plot(x, y_shanghai, color="b", linestyle="--", label="上海")
axes[1].plot(x, y_beijin, color="r", linestyle=":", label="北京")


# 显示图例
axes[0].legend(loc="best")
axes[1].legend(loc="best")
# 修改x，y刻度
# 准备x的刻度说明
x_label = ["11点{}分".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(0, 30, 2))

axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(0, 30, 2))
# 添加网格显示
axes[0].legend(loc="best")
axes[1].legend(loc="best")

# 添加描述信息
axes[0].set_xlabel("时间变化")
axes[0].set_ylabel("温度变化")
axes[0].set_title("上海11-12点的温度变化")

axes[1].set_xlabel("时间变化")
axes[1].set_ylabel("温度变化")
axes[1].set_title("北京11-12点的温度变化")
# 4.显示图像
plt.show()
```

## 柱状图绘制

```python
# 需求-对比每部电影的票房收入
# 1.准备数据
x = ['雷神3：诸神黄昏', '正义联盟', '东方快车谋杀案', '寻梦环游记',
     '全球风暴', '降魔传', '追捕', '七十七天', '密战', '狂兽', '其它']
y = [73853, 57767, 22354, 15969, 14839, 8725, 8716, 8318, 7916, 6764, 52222]

# 2.创建画布
plt.rc('font', family='simsun', size=12)
plt.figure(figsize=(20, 8), dpi=80)

# 3.绘制柱状图
plt.bar(x, y, width=0.5, color=['b', 'r', 'g',
        'y', 'c', 'm', 'y', 'k', 'c', 'g', 'b'])
# 添加标题
plt.title("电影的票房收入")
# 修改x轴的刻度显示
plt.xticks(range(len(x)), x)
# 添加网格显示
plt.grid(linestyle="--", alpha=0.2)
# 4.显示图像
plt.show()
```

```python
# 需求2-对比电影票房收入才能更有说服力
#  1.准备数据
movie = ['雷神3：诸神黄昏', '正义联盟', '寻梦环游记']
first_day = [10587.6, 10062.5, 1275.7]
first_weekend = [36224.9, 34479.6, 11830]
# 2.创建画布
plt.rc('font', family='simsun', size=12)
plt.figure(figsize=(20, 8), dpi=80)

# 3.绘制柱状图
plt.bar(range(3), first_day, width=0.2, label="首日票房")
plt.bar([i+0.21 for i in range(3)], first_weekend, width=0.2, label="首周票房")
# 添加标题
plt.title("电影的票房收入对比")
plt.xlabel("电影类别")
plt.ylabel("票房")
plt.legend()
# 修改x轴的刻度显示
plt.xticks([i+0.1 for i in range(3)], movie)


# 添加网格显示
plt.grid(linestyle="--", alpha=0.2)
# 4.显示图像
plt.show()
```

## 直方图绘制


```python
# 需求：电影时长分布状况
import numpy as np
# 1.准备数据
time = [131, 98, 125, 131, 124, 139, 131, 117, 128, 108,
        135, 138, 131, 12, 12, 107, 12, 12, 12, 12, 114]
distance = 3
bins = (max(time) - min(time)) // distance  # 计算组数

# 2.创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3.绘制直方图
plt.hist(time, bins, density=True)
plt.xticks(np.linspace(min(time), max(time),bins+1,dtype=np.int32))

# 添加网格
plt.grid(linestyle="--", alpha=0.25)

# 4.显示图像
plt.show()
```

## 饼图绘制

```python
#
# #需求：电影时长分布状况

#设置中文字体 
plt.rc('font', family='simsun', size=13)



# 1.准备数据
movie_name = ['雷神3：诸神黄昏', '正义联盟', '东方快车谋杀案', '寻梦环游记',
              '全球风暴', '降魔传', '追捕', '七十七天', '密战', '狂兽', '其它']
place_count = [73853, 57767, 22354, 15969,
               14839, 8725, 8716, 8318, 7916, 6764, 52222]

# 2.创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3.绘制直方图
# autopct 显示百分比
plt.pie(place_count, labels=movie_name, colors=[
        'b', 'r', 'g', 'y', 'c', 'm', 'y', 'k', 'c', 'g', 'b'], autopct="%1.2f%%")
# 显示图例
plt.legend()
# 将椭圆变成圆
plt.axis("equal")
# 4.显示图像
plt.show()
```
