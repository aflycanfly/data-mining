```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

# 读取数据
coder = pd.read_excel('./word_data/bianma.xlsx')
coder.iloc[:6,:12]
```

```python
# 设置背景风格 虚线网格
sns.set_style("whitegrid",{"grid.alpha": "0.1", "grid.linestyle": "--"})
# 设置显示中文
mpl.rcParams['font.sans-serif'] = ['SimHei']
# 设置x轴 字体大小
mpl.rcParams['xtick.labelsize'] = 8

# 设置每个柱子宽度shrink=0.8 柱子边框颜色edgecolor=".3" 柱子边框宽度linewidth=0.7 柱子透明度alpha=0.9 内容主题multiple="stack" A1-A6 叠放
sns.histplot(data=coder, x="作者", hue="内容主题", multiple="stack",shrink=0.8,edgecolor=".3",linewidth=0.7,alpha=0.9)

plt.ylim(0, 13)
plt.xticks(rotation=45)# x轴坐标旋转45度

plt.xlabel('作者')
plt.ylabel('数量')
plt.title('作者与内容主题的关系')
# 设置图例位置 legend
plt.legend(labels = ['A1','A2','A3','A4','A5','A6'],loc = 2, bbox_to_anchor = (1,1))

```
