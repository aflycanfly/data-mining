####  如何创建DataFrame

```python
import pandas as pd
import numpy as np
```

```python
stock_change = np.random.normal(loc=1, scale=0.5, size=(10, 8))
```

```python
pd.DataFrame(stock_change)
```

```python
# 添加列索引
stock_name = ["股票{}".format(i) for i in range(1, 11)]
stock_name
```

```python
pd.DataFrame(stock_change, index=stock_name)
```

```python
# 添加行索引
stock_colum = pd.date_range(start="20200101", periods=8)
```

```python
stock_colum
```

```python
data = pd.DataFrame(stock_change, index=stock_name, columns=stock_colum)
data
```

### 基本数据操作

#### 索引操作

```python
data = pd.read_csv("../../resource/data-mining/stock_day.csv")
```

```python
data
```

```python
# 删除一些列，让数据更简单些，再去做后面的操作
data = data.drop(["ma5", "ma10", "ma20", "v_ma5", "v_ma10", "v_ma20"], axis=1)
```

```python
data
```

```python
# 1.直接索引
# data[1, 0]  不能进行数字索引
data["open"]["2018-02-27"]  # 必须先列后行
```

```python
# 2.按名字索引
data.loc["2018-02-27"]["open"]
```

```python
data.loc["2018-02-27", "open"]
```

```python
# 3.按数字索引
data.iloc[1, 0]
```

```python
data.iloc[1:, 0:]
```

```python
data.iloc[1][0]
```

```python
# 4.组合索引
data.loc[data.index[0:4], ["open", "close", "high", "low"]]
```

```python
data.iloc[0:4, data.columns.get_indexer(["open", "close", "high", "low"])]
```

#### 赋值操作

```python
data.open = 100
data
```

```python
data.iloc[:1, :] = 100
```

```python
data.head()
```

#### 排序

```python
data.sort_values(by="high", ascending=False)  # 以high字段排序，False降序
```

```python
# 先按high排序，后按price_change排序
data.sort_values(by=["high", "price_change"], ascending=False)
```

```python
data.sort_index()  # 按索引进行排序
```

```python
ser = data["high"]
```

```python
ser.sort_values()  # 按值进行排序
```

```python
ser.sort_index(ascending=False)
```

#### 逻辑运算

##### 逻辑运算符

```python
# 获取p_change > 2 的日期
data[data["p_change"] > 2].head()  # 做布尔索引
```

```python
data["p_change"] > 2
```

```python
# 筛选p_change>2 并且 low>15
data[(data["p_change"] > 2) & (data["low"] > 15)]
```

##### 逻辑运算函数

```python
data.query("p_change>2 & low > 15")
```

```python
# 判断turnover 是否为4.19,2.39
data[data["turnover"].isin([4.19, 2.39])]
```

#### 统计运算

##### 获取综合的统计指标

```python
# 获取综合统计指标
data.describe().sort_values("volume")
```

##### 获取具体的统计指标

```python
# 注意axis的设置
data.max(axis=0) #按列去获得最大值 默认按列去获取最大值
```

```python
data.max(axis=1).sort_values().head()    #按行去获得最大值
```

```python
data.idxmax()   #索引每一列的最大值都在哪一行
```

```python
data.sum()
```

```python
data.mode()
```

```python
data.abs()
```
