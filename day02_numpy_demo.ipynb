## ndarray与Python原生list运算效率对比

```python
import random
import time
```

```python
# 生成一个大的数组
python_list = []
for i in range(100000000):
    python_list.append(i)
```

```python
ndarray_list = np.array(python_list)
```

```python
# python 原生数组求和
t1 = time.time()
python_sum = sum(python_list)
t2 = time.time()
d1 = t2-t1
# ndarray 求和
t1 = time.time()
ndarray_sum = np.sum(ndarray_list)
t2 = time.time()
d2 = t2-t1
print("时间相差", d1, d2)
```

### ndarray运算

#### 逻辑运算

```python
# 生成股票股势
stock_change = np.random.normal(loc=0, scale=0.1, size=(8, 10))
stock_change
```

```python
# 逻辑判断，如果涨跌幅大于0.5就标记为True 否则False
stock_change > 0.5
```

```python
# 布尔索引 保留涨跌幅大于0.5
stock_change[stock_change > 0.5]
```

```python
# 布尔索引 涨跌幅大于0.5的改成1.1
stock_change[stock_change > 0.5] = 0
stock_change
```

#### 通用判断函数

```python
# 判断stock_change[0:2,0:5]是否全是上涨的
stock_change[0:2, 0:5] > 0
```

```python
np.all(stock_change[0:2, 0:5] > 0)  # 有一个是False就是False了
```

```python
# 判断前五行这一段时间是否有上涨
np.any(stock_change[:5, :] > 0)  # 有一个是False就是False了
```

#### np.where（三元运算符）

```python
# 判断前四个股票的前四天的涨跌幅，大于0的置为1，否则置为0
temp = stock_change[:4, :]
```

```python
np.where(temp > 0, 1, 0)
```

```python
# 判断前四天股票的涨跌幅,大于0并且小于0.1的，换为1，否则换为0
np.logical_and(temp > 0, temp < 0.1)
```

```python
np.where(np.logical_and(temp > 0, temp < 0.1), 1, 0)
```

```python
# 判断前四天股票的涨跌幅,大于0.1或者小于-0.1的，换为1，否则换为0
np.logical_or(temp > 0.1, temp < -0.1)
```

```python
np.where(np.logical_or(temp > 0.1, temp < -0.1), 1, 0)
```

#### 矩阵乘法

```python
# 用ndarray存储
score = np.array([[80, 86],
                  [82, 80],
                  [85, 78],
                  [90, 90],
                  [86, 82],
                  [82, 90],
                  [78, 80],
                  [92, 94]])
weights = np.array([[0.7], [0.3]])
```

```python
np.matmul(score,weights) #矩阵相乘
```

```python
np.dot(score,weights) #矩阵相乘
```

```python
score @ weights
```

```python
# 用matrix存储
score_mat = np.mat(score)
weights_mat = np.mat(weights)
```

```python
score_mat*weights_mat
```
